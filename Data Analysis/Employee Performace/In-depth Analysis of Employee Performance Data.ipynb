```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'Employees.xlsx'
employees_df = pd.read_excel(file_path, sheet_name='Employees')
```

```python
# Set up the visual style
sns.set(style="whitegrid")
```

```python
# 1. Gender Distribution of Employees
plt.figure(figsize=(8, 6))
sns.countplot(data=employees_df, x='Gender', palette='pastel')
plt.title('Gender Distribution of Employees')
plt.ylabel('Number of Employees')
plt.xlabel('Gender')
plt.show()
```

```python
# 2. Salary Distribution by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=employees_df, x='Gender', y='Monthly Salary', palette='pastel')
plt.title('Monthly Salary Distribution by Gender')
plt.ylabel('Monthly Salary')
plt.xlabel('Gender')
plt.show()
```

```python
# 3. Salary Distribution by Department and Gender
plt.figure(figsize=(12, 6))
sns.boxplot(data=employees_df, x='Department', y='Monthly Salary', hue='Gender', palette='pastel')
plt.title('Monthly Salary Distribution by Department and Gender')
plt.ylabel('Monthly Salary')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Salary Progression by Years of Service and Gender
plt.figure(figsize=(10, 6))
sns.lineplot(data=employees_df, x='Years', y='Monthly Salary', hue='Gender', marker='o', palette='pastel')
plt.title('Salary Progression by Years of Service and Gender')
plt.ylabel('Monthly Salary')
plt.xlabel('Years of Service')
plt.show()
```

```python
# 5. Department Representation by Gender
plt.figure(figsize=(12, 6))
sns.countplot(data=employees_df, x='Department', hue='Gender', palette='pastel')
plt.title('Department Representation by Gender')
plt.ylabel('Number of Employees')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Overtime Hours by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=employees_df, x='Gender', y='Overtime Hours', palette='pastel')
plt.title('Overtime Hours by Gender')
plt.ylabel('Overtime Hours')
plt.xlabel('Gender')
plt.show()
```

```python
# 7. Sick and Unpaid Leaves by Gender
plt.figure(figsize=(14, 6))

# Sick Leaves
plt.subplot(1, 2, 1)
sns.boxplot(data=employees_df, x='Gender', y='Sick Leaves', palette='pastel')
plt.title('Sick Leaves by Gender')
plt.ylabel('Number of Sick Leaves')
plt.xlabel('Gender')

# Unpaid Leaves
plt.subplot(1, 2, 2)
sns.boxplot(data=employees_df, x='Gender', y='Unpaid Leaves', palette='pastel')
plt.title('Unpaid Leaves by Gender')
plt.ylabel('Number of Unpaid Leaves')
plt.xlabel('Gender')

plt.tight_layout()
plt.show()
```
