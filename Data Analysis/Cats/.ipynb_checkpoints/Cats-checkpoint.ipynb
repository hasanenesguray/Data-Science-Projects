# Let's predict the Breed

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ['Breed', 'Color','Gender']

# Initialize LabelEncoder
encoder = LabelEncoder()

# Loop through each column to encode
for column in columns_to_encode:
    df[column + '_Encoded'] = encoder.fit_transform(df[column])
    
df.head()
df.info()
```

```python
df_encoded = df.loc[:, ['Age_In_Years','Weight_In_Kg','Color_Encoded','Gender_Encoded','Breed_Encoded']]
df_encoded.corr()
```

```python
X = df_encoded.loc[:, ['Age_In_Years','Weight_In_Kg','Color_Encoded','Gender_Encoded','Breed_Encoded']]
y= df.loc[:, ['Breed_Encoded']]
X.head()
```

```python
y.head()
y.info()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
y = np.ravel(y)
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg_param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs'],
    'penalty': ['l2']  # Note: 'l1' penalty is only supported with 'liblinear' solver
}

# Grid Search for Logistic Regression
grid_search_log_reg = GridSearchCV(estimator=log_reg, 
                                    param_grid=log_reg_param_grid, 
                                    cv=5, 
                                    scoring='accuracy', 
                                    verbose=1, 
                                    n_jobs=-1)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Fit Grid Search for Logistic Regression
grid_search_log_reg.fit(X_train_scaled, y_train)


# Get the best parameters and the best model for Logistic Regression
best_params_log_reg = grid_search_log_reg.best_params_
best_model_log_reg = grid_search_log_reg.best_estimator_

print(f"Best Parameters for Logistic Regression: {best_params_log_reg}")
y_pred_log_reg = best_model_log_reg.predict(X_test_scaled)
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_log_reg))
```
