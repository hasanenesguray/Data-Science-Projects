```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'Amazon Sale Report.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data.head()
```

```python
# Summary of the data to get a better understanding of the data types and any missing values
data_summary = data.describe(include='all')
data_info = data.info()

# Checking for missing values
missing_values = data.isnull().sum()

# Display summary, info, and missing values
data_summary, missing_values
```

* The Amount column shows a mean value of 648.56 INR, with a range from 0 to 5,584 INR.
* Most orders are shipped to major cities in India, with Maharashtra being the most common state.

# Order Status Distribution

```python
# Visualizing Order Status Distribution
plt.figure(figsize=(10, 6))
data['Status'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha="right")
plt.show()
```

* The majority of the orders have a status of "Shipped," indicating successful transactions.
* Other statuses like "Cancelled" and "Delivered to Buyer" are also significant, showing various stages of the order process.

# Fulfillment Method Distribution

```python
# Visualizing Fulfilment Method Distribution
plt.figure(figsize=(10, 6))
data['Fulfilment'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Fulfilment Method Distribution')
plt.ylabel('')
plt.show()
```

* The pie chart shows that a higher percentage of orders are fulfilled by Amazon compared to merchants, indicating a preference or reliance on Amazon's fulfillment services.

# Amount Distribution

```python
# Visualizing Amount Distribution
plt.figure(figsize=(10, 6))
data['Amount'].dropna().plot(kind='hist', bins=50, color='orange', edgecolor='black')
plt.title('Amount Distribution')
plt.xlabel('Amount (INR)')
plt.ylabel('Frequency')
plt.show()
```

* The histogram reveals that most orders are in the lower price range, with a significant concentration around 500 to 1000 INR. This suggests that the majority of products sold are relatively low-cost items.

# Top 10 Shipping States

```python
# Visualizing Shipping State Distribution
plt.figure(figsize=(10, 6))
data['ship-state'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Shipping States')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha="right")
plt.show()
```

* Maharashtra, Karnataka, and Tamil Nadu are the top shipping destinations, reflecting a high concentration of orders in these regions.

# Threshold for high-value orders

```python
# Define a threshold for high-value orders. For this example, let's consider orders above 1000 INR as high-value.
high_value_threshold = 1000

# Filter the dataset to include only high-value orders
high_value_orders = data[data['Amount'] > high_value_threshold]

# Display the number of high-value orders and their summary
high_value_summary = high_value_orders.describe(include='all')

high_value_orders.shape, high_value_summary
```

The filtered dataset contains 13,327 high-value orders (orders above 1000 INR). Here are some key details:

* Order Status: The majority of these orders have a status of "Shipped," indicating successful transactions.

* Fulfillment: Most of these high-value orders were fulfilled by Amazon, with a significant number being handled by the "Expedited" shipping service.

* Geographical Distribution: Maharashtra continues to be the top state for high-value orders, with Bengaluru as the most common city.

# Different shipping methods

```python
# Analyzing the distribution of different shipping methods
shipping_method_distribution = data['ship-service-level'].value_counts()

# Visualizing the shipping method distribution
plt.figure(figsize=(10, 6))
shipping_method_distribution.plot(kind='bar', color='teal')
plt.title('Shipping Method Distribution')
plt.xlabel('Shipping Method')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Shipping Method Distribution:

* The majority of orders used the "Expedited" shipping method, with 88,630 orders.
* The "Standard" shipping method was used for 40,346 orders.

# Relationship between shipping method and order amount

```python
# Further analysis: relationship between shipping method and order amount
shipping_vs_amount = data.groupby('ship-service-level')['Amount'].mean()

# Visualizing the average order amount by shipping method
plt.figure(figsize=(10, 6))
shipping_vs_amount.plot(kind='bar', color='orange')
plt.title('Average Order Amount by Shipping Method')
plt.xlabel('Shipping Method')
plt.ylabel('Average Amount (INR)')
plt.xticks(rotation=45, ha="right")
plt.show()

shipping_method_distribution, shipping_vs_amount
```

## Average Order Amount by Shipping Method:

* Orders with "Expedited" shipping had a slightly higher average order amount (656.22 INR) compared to "Standard" shipping (632.08 INR).
* This suggests that customers opting for faster delivery (Expedited) tend to place higher-value orders.

# Orders and shipping method analysis

```python
# Saving the high-value orders and shipping method analysis to a new CSV file
output_path = 'Amazon_Sale_High_Value_Orders_Analysis.xlsx'

# Creating a summary DataFrame for shipping method analysis
shipping_analysis_df = pd.DataFrame({
    'Shipping Method': shipping_vs_amount.index,
    'Number of Orders': shipping_method_distribution.values,
    'Average Order Amount (INR)': shipping_vs_amount.values
})

# Saving both high-value orders and shipping method analysis to the Excel file
with pd.ExcelWriter(output_path) as writer:
    high_value_orders.to_excel(writer, sheet_name='High Value Orders', index=False)
    shipping_analysis_df.to_excel(writer, sheet_name='Shipping Method Analysis', index=False)

output_path
```
