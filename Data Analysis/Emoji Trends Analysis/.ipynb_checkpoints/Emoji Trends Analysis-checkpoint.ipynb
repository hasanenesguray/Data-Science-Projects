```python
import pandas as pd

# Load the dataset
df_emoji = pd.read_csv('emoji_usage_dataset.csv')

# Display the first few rows and column names
print("Columns in the dataset:")
print(df_emoji.columns)
print("\nFirst few rows of the dataset:")
print(df_emoji.head())
```

```python
# Summary statistics for numerical columns
print("Summary statistics for numerical columns:")
print(df_emoji[['User Age']].describe())

# Summary statistics for categorical columns
print("\nSummary statistics for categorical columns:")
print(df_emoji[['Emoji', 'Context', 'Platform', 'User Gender']].describe())
```

```python
import matplotlib.pyplot as plt

# Count occurrences of each emoji
emoji_counts = df_emoji['Emoji'].value_counts()

# Plot the distribution of emojis
plt.figure(figsize=(12, 8))
emoji_counts.plot(kind='bar', color=plt.get_cmap('Set1').colors)
plt.title('Distribution of Emojis')
plt.xlabel('Emoji')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
# Plot the distribution of user ages
plt.figure(figsize=(12, 8))
plt.hist(df_emoji['User Age'].dropna(), bins=20, color='lightblue', edgecolor='black')
plt.title('Distribution of User Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Count occurrences of each context
context_counts = df_emoji['Context'].value_counts()

# Plot the distribution of contexts
plt.figure(figsize=(12, 8))
context_counts.plot(kind='bar', color=plt.get_cmap('Pastel1').colors)
plt.title('Distribution of Contexts')
plt.xlabel('Context')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
# Count occurrences of each platform
platform_counts = df_emoji['Platform'].value_counts()

# Plot the distribution of platforms
plt.figure(figsize=(12, 8))
platform_counts.plot(kind='bar', color=plt.get_cmap('Paired').colors)
plt.title('Distribution of Platforms')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
# Count occurrences of each gender
gender_counts = df_emoji['User Gender'].value_counts()

# Plot the distribution of user genders
plt.figure(figsize=(8, 8))
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.get_cmap('Set2').colors)
plt.title('Distribution of User Genders')
plt.ylabel('')
plt.show()
```

```python
# Check for missing values in the dataset
print("Missing values in each column:")
print(df_emoji.isnull().sum())
```

```python
import seaborn as sns

# Since there are no other numerical columns except 'User Age', correlation matrix will be limited
correlation_matrix = df_emoji[['User Age']].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

```python
# Boxplot of Age by Emoji
plt.figure(figsize=(14, 10))
sns.boxplot(x='Emoji', y='User Age', data=df_emoji, palette='tab20')
plt.xticks(rotation=45, ha='right')
plt.title('Age Distribution by Emoji')
plt.xlabel('Emoji')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Age by Context
plt.figure(figsize=(14, 10))
sns.boxplot(x='Context', y='User Age', data=df_emoji, palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.title('Age Distribution by Context')
plt.xlabel('Context')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Age by Platform
plt.figure(figsize=(14, 10))
sns.boxplot(x='Platform', y='User Age', data=df_emoji, palette='Set1')
plt.xticks(rotation=45, ha='right')
plt.title('Age Distribution by Platform')
plt.xlabel('Platform')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Age by Gender
plt.figure(figsize=(12, 8))
sns.boxplot(x='User Gender', y='User Age', data=df_emoji, palette='pastel')
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Count occurrences of each emoji by platform
emoji_platform_counts = df_emoji.groupby('Platform')['Emoji'].value_counts().unstack().fillna(0)

# Plot the emoji usage by platform
plt.figure(figsize=(14, 10))
emoji_platform_counts.plot(kind='bar', stacked=True, colormap='tab20')
plt.title('Emoji Usage by Platform')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Emoji')
plt.grid(axis='y')
plt.show()
```

```python
# Count occurrences of each emoji by context
emoji_context_counts = df_emoji.groupby('Context')['Emoji'].value_counts().unstack().fillna(0)

# Plot the emoji usage by context
plt.figure(figsize=(14, 10))
emoji_context_counts.plot(kind='bar', stacked=True, colormap='tab20')
plt.title('Emoji Usage by Context')
plt.xlabel('Context')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Emoji')
plt.grid(axis='y')
plt.show()
```
