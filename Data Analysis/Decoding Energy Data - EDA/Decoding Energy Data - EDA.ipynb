```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#load data
data = pd.read_csv('complete_renewable_energy_dataset.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
#Percentage of missing values in the dataset
missing_percentage = (data.isnull().sum() / len(data)) * 100
print(missing_percentage)
```

# EDA

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=data, x='Country', y='Production (GWh)')
plt.title('Distribution of Energy Production by Country')
plt.xticks(rotation=90)
plt.ylabel('Production (GWh)')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=data, x='Year', y='Production (GWh)', hue='Energy Type')
plt.title('Energy Production Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Production (GWh)')
plt.legend(title='Energy Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=data, x='Country', y='Investments (USD)', ci=None)
plt.title('Investment in Renewable Energy by Country')
plt.xticks(rotation=90)
plt.ylabel('Investments (USD)')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of All Features')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=data, x='Country', y='Proportion of Energy from Renewables')
plt.title('Proportion of Energy from Renewables by Country')
plt.xticks(rotation=90)
plt.ylabel('Proportion of Energy from Renewables (%)')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=data, x='GDP', y='Energy Consumption', hue='Country', size='Population', sizes=(20, 200))
plt.title('GDP vs Energy Consumption')
plt.xlabel('GDP')
plt.ylabel('Energy Consumption')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=data, x='Energy Type')
plt.title('Distribution of Energy Types')
plt.xlabel('Energy Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(data=data, x='Innovation Index', y='Ease of Doing Business', hue='Country', style='Energy Type', size='GDP', sizes=(20, 200))
plt.title('Innovation Index vs Ease of Doing Business')
plt.xlabel('Innovation Index')
plt.ylabel('Ease of Doing Business')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
