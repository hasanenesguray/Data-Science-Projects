Data Preprocessing

```python
from datetime import datetime
```

```python
today = datetime.now()
```

```python
year_work = []
for i in df['work_year']:
  year_work.append(today.year - i)

df['work_year'] = year_work
```

Here, I convert work year to experience of work to use better in machine learning algorithms

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df['experience_level'].unique()
```

For experience level:

EX -> Executive

SE -> Senior

MI -> Middle

EN -> Entry Level

```python
df.employment_type.unique()
```

FT -> Full Time

CT -> Contract

PT -> Part Time

FL -> Freelance

```python
df.head(3)
```

```python
label_encoder = LabelEncoder()
#df['experience_level'] = label_encoder.fit_transform(df['experience_level'])
#df['employment_type'] = label_encoder.fit_transform(df['employment_type'])
#df['company_size'] = label_encoder.fit_transform(df['company_size'])
df['job_title'] = label_encoder.fit_transform(df['job_title'])
df['company_location'] = label_encoder.fit_transform(df['company_location'])
df['employee_residence'] = label_encoder.fit_transform(df['employee_residence'])
```

In label encoding, object type values converts to some integer numbers due to a hierarchy

```python
df = pd.get_dummies(df, columns = ['company_size', 'employment_type', 'experience_level'], drop_first = True)
```

I want to see all salaries in the same currency so I will just use 'salary_in_usd' column so I do not need salary and salary_currency columns.

```python
df = df.drop(['salary', 'salary_currency'], axis = 1)
```
