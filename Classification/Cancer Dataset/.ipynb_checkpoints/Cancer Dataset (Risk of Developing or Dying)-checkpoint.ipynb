# Problem Statement

## Title
Comparative Analysis of Gender-Specific Cancer Risk and Mortality Rates: Insights from Development and Dying Ratios Across Different Cancer Types

## Objective
This research aims to analyze and compare the cancer risk and mortality rates between males and females across different cancer types. By examining both development and dying ratios, the study seeks to provide insights into gender-specific cancer risks and outcomes, highlighting key differences and trends.

## Dataset Description
The dataset contains the following columns:

* Cancer Type: The type of cancer being analyzed (e.g., Bladder, Breast, Brain and Nervous System).
* Male Risk Development Percentage: The percentage chance that a male will develop the specified cancer over their lifetime.
* Male Risk Development Ratio (One in __ Person): The likelihood ratio for males developing the cancer compared to the general population.
* Male Risk Dying Percentage: The percentage chance that a male will die from the cancer if diagnosed.
* Male Risk Dying Ratio (One in __ Person): The likelihood ratio for males dying from the cancer compared to the general population.
* Female Risk Development Percentage: The percentage chance that a female will develop the specified cancer over their lifetime.
* Female Risk Development Ratio (One in __ Person): The likelihood ratio for females developing the cancer compared to the general population.
* Female Risk Dying Percentage: The percentage chance that a female will die from the cancer if diagnosed.
* Female Risk Dying Ratio (One in __ Person): The likelihood ratio for females dying from the cancer compared to the general population.

## 3. Exploratory Data Analysis (EDA)

### 3.1 Univariate Analysis
Univariate analysis focuses on examining the distribution and characteristics of individual variables in the dataset. This analysis provides insights into the distribution, central tendency, and variability of each feature.

```python
# Summary statistics
data.describe(include='all')
```

```python
# Distribution of Cancer Types
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Cancer Type', palette='viridis')
plt.xticks(rotation=90)
plt.title('Distribution of Cancer Types')
plt.show()
```

```python
# Distribution of Male and Female Risk Development Percentages
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(data['Male Risk Development Percentage'], kde=True, color='blue', ax=axes[0])
axes[0].set_title('Distribution of Male Risk Development Percentage')

sns.histplot(data['Female Risk Development Percentage'], kde=True, color='red', ax=axes[1])
axes[1].set_title('Distribution of Female Risk Development Percentage')

plt.show()
```

### 3.2 Bivariate Analysis

```python
# Relationship between Male and Female Risk Development Percentages
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Male Risk Development Percentage', y='Female Risk Development Percentage', hue='Cancer Type')
plt.title('Male vs. Female Risk Development Percentages')
plt.show()
```

```python
# Relationship between Male and Female Risk Dying Percentages
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Male Risk Dying Percentage', y='Female Risk Dying Percentage', hue='Cancer Type')
plt.title('Male vs. Female Risk Dying Percentages')
plt.show()
```

### 3.3 Advanced Visualization

```python
# Violin Plot for Risk Development Percentages by Cancer Type
plt.figure(figsize=(14, 7))
sns.violinplot(data=data, x='Cancer Type', y='Male Risk Development Percentage', palette='viridis')
plt.xticks(rotation=90)
plt.title('Violin Plot of Male Risk Development Percentage by Cancer Type')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.violinplot(data=data, x='Cancer Type', y='Female Risk Development Percentage', palette='viridis')
plt.xticks(rotation=90)
plt.title('Violin Plot of Female Risk Development Percentage by Cancer Type')
plt.show()
```

```python
# Box Plot for Risk Dying Percentages by Cancer Type
plt.figure(figsize=(14, 7))
sns.boxplot(data=data, x='Cancer Type', y='Male Risk Dying Percentage', palette='viridis')
plt.xticks(rotation=90)
plt.title('Box Plot of Male Risk Dying Percentage by Cancer Type')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=data, x='Cancer Type', y='Female Risk Dying Percentage', palette='viridis')
plt.xticks(rotation=90)
plt.title('Box Plot of Female Risk Dying Percentage by Cancer Type')
plt.show()
```

### 3.4 Correlation Analysis

```python
# Correlation Matrix
plt.figure(figsize=(12, 8))
correlation_matrix = data[['Male Risk Development Percentage', 'Female Risk Development Percentage',
                           'Male Risk Dying Percentage', 'Female Risk Dying Percentage']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
